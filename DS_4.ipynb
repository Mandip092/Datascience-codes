```python
import numpy as np
import matplotlib.pyplot as plt

n = 10000
p = 0.7

X = np.random.binomial(1,p,n)

cumulative_avg = np.cumsum(X) / np.arange(1,n+1)

plt.figure(figsize=(10,6))
plt.plot(cumulative_avg, label="Cumulative Average (A_n)")
plt.axhline(y=p, linestyle='--', label="True Probability (p = 0.7)")

plt.xlabel("Number of Trials (n)")
plt.ylabel("Cumulative Average")
plt.title("Law of Large Numbers (Bernoulli Trials)")
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

m = 1000
k_values = [1,2,5,30]

plt.figure(figsize=(20,6))
for i, k in enumerate(k_values, 1):
  samples = np.random.randint(1, 7, size=(m, k))
  sample_means = np.mean(samples, axis=1)
  plt.subplot(1, 4, i)
  plt.hist(sample_means, bins=60, density=True,  alpha=0.6, color='green')
  plt.title(f"k = {k}")
  plt.xlabel("Sample Mean")
  plt.ylabel("Density")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

lambda_val = 0.5
sample_size = 50
num_samples = 500

sample_means = []

for _ in range(num_samples):
  sample = np.random.exponential(scale=1/lambda_val, size=sample_size)
  sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

mu = 1 / lambda_val
sigma = (1 / lambda_val) / np.sqrt(sample_size)

plt.figure(figsize=(6,6))
stats.probplot(sample_means, dist="norm", sparams=(mu, sigma), plot=plt)

plt.title("Q-Q Plot of Sample Means vs Normal Distribution")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

levels = 100
p_right = 0.5
num_balls = 5000

fp = np.zeros((100, num_balls))

for i in range(100):
  trials = np.random.binomial(1, p_right, size=(num_balls, levels))
  fp[i] = np.sum(trials, axis=1)

final_positions = np.mean(fp, axis=0)

plt.figure(figsize=(8,5))
count, bins, ignored = plt.hist(final_positions, bins=30, density=True)

mu = levels * p_right
sigma = np.sqrt(levels * p_right * (1 - p_right)) / np.sqrt(100)

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 200)
plt.plot(x, stats.norm.pdf(x, mu, sigma), linewidth=2)

plt.title("Digital Galton Board: Binomial vs. Normal Distribution")
plt.xlabel("Final Position (Number of Right Moves)")
plt.ylabel("Probability Density")
plt.grid()
plt.show()
```
