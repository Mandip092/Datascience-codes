```python
def is_even(n):
  if n % 2 == 0:
    return True
  else:
    return False

num = int(input("Enter a number: "))

if is_even(num):
  print("True")
else:
  print("False")
```

```python
n = int(input("Enter number of elements: "))
numbers = []

for i in range(n):
  numbers.append(int(input()))

  even_numbers = []
  for num in numbers:
    if num % 2 == 0:
      even_numbers.append(num)

print("Original List: ", numbers)
print("Even Numbers List: ", even_numbers)
```

```python
#student_marks_analyzer.py

n = int(input("Enter the number of students: "))

marks =[]

for i in range(n):
  mark = int(input(f"Enter the marks for student {i+1} (0-100): "))
  marks.append(mark)

print("\nList of all marks: ", marks)

total_marks = sum(marks)
print("Total of all marks: ", total_marks)

average_marks = total_marks / n
print("Average marks: ", average_marks)

print("Highest marks: ", max(marks))
print("Lowest marks: ", min(marks))

passed = 0
failed = 0

for mark in marks:
  if mark >= 40:
    passed += 1
  else:
    failed += 1

print("Number of students passed: ", passed)
print("Number of students failed: ", failed)

print("Marks in reverse order: ", marks[::-1])

even_marks = []
for mark in marks:
  if mark % 2 == 0:
    even_marks.append(mark)

print("Even marks: ", even_marks)
```
