```python
import numpy as np
import matplotlib.pyplot as plt

num_customers = 1000   # number of independent customers
purchase_prob = 0.30   # probability that a customer makes a purchase

# Each customer outcome:
# 1 = purchase, 0 = no purchase
customer_outcomes = np.random.binomial(
    n=1,
    p=purchase_prob,
    size=num_customers
)

mean_outcome = np.mean(customer_outcomes)
variance_outcome = np.var(customer_outcomes)

print("Simulation Results")
print("------------------")
print(f"Number of customers: {num_customers}")
print(f"Probability of purchase: {purchase_prob}")
print(f"Mean (Expected value): {mean_outcome}")
print(f"Variance: {variance_outcome}")
print(f"Total purchases: {np.sum(customer_outcomes)}")

values, counts = np.unique(customer_outcomes, return_counts=True)

plt.figure(figsize=(6, 4))
plt.bar(values, counts, color=["skyblue", "green"], edgecolor="black")
plt.xticks([0, 1], ["No Purchase (0)", "Purchase (1)"])
plt.ylabel("Number of Customers")
plt.xlabel("Customer Outcome")
plt.title("Customer Purchase Decisions (1000 Customers)")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 50        # fixed number of customers
p = 0.6       # probability that a customer makes a purchase

print("Experiment Structure:")
print("Number of trials (n):", n)
print("Each trial has two outcomes: Purchase (Success) or No Purchase (Failure)")
print("Probability of success (p):", p)
print("Trials are independent.\n")

print("Distribution Used: Binomial Distribution\n")


x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)


plt.figure(figsize=(10,6))
plt.bar(x, pmf, color='skyblue')
plt.title("Binomial PMF (n=50, p=0.6)")
plt.xlabel("Number of Purchasing Customers")
plt.ylabel("Probability")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


most_likely = x[np.argmax(pmf)]
print("Most likely number of purchasing customers:", most_likely)
print("This value represents the mode of the Binomial distribution.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lam = 10

p_8 = poisson.pmf(8, lam)
print("P(X = 8):", p_8)

p_more_15 = 1 - poisson.cdf(15, lam)
print("P(X > 15):", p_more_15)

x = np.arange(0, 26)
pmf = poisson.pmf(x, lam)

plt.figure(figsize=(10, 5))
plt.bar(x, pmf, color='lightgreen')
plt.title('Probability Mass Function of Checkouts')
plt.xlabel('Number of Checkouts per Hour')
plt.ylabel('Probability Mass Function')
plt.xticks(x)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

# Probability of success
p = 0.25

# Define range for first 10 interactions
k = np.arange(1, 11)

# Geometric PMF
pmf = geom.pmf(k, p)

# Probability that first success occurs on 4th interaction
prob_4th = geom.pmf(4, p)

print("Probability that first success occurs on 4th interaction:")
print(prob_4th)

# Plot PMF
plt.figure(figsize=(8,5))
plt.stem(k, pmf)
plt.title("PMF of Geometric Distribution (p = 0.25)")
plt.xlabel("Number of Interactions Until First Success")
plt.ylabel("Probability")
plt.xticks(k)
plt.grid(True)
plt.show()
```
